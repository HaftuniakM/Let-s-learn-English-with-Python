```python
# Downloading Python libraries
import random   
import matplotlib.pyplot as plt  
import pandas as pd

# Read csv file from GitHub into DataFrame:
url = 'https://raw.githubusercontent.com/HaftuniakM/Let-started-learning-English-with-Python/main/List_of_irregular_Eng_verbs.csv'
df = pd.read_csv(url)
df.head(3)
```

```python
# Creating keys and values for dictionaries:
keys = list(df['Infinitive'])
values_2form = list(df['Simple Past'])
values_3form = list(df['Past participle'])

# Creating dictionaries of the second and third forms:
res_2form = {keys[i]: values_2form[i] for i in range(len(keys))}
res_3form = {keys[i]: values_3form[i] for i in range(len(keys))}
```

```python
# CHECK 2 form of irregular verbs
total = 0
right_answers = 0

while True:
    random_verb = random.choice(keys)
    input_verb = input(f'Please enter the verb in form 2 for {random_verb}')
    if input_verb == res_2form[random_verb]:
        right_answers += 1
        print('Ok, you are right!', f'{random_verb} --> {input_verb}', sep = '\n')
    elif input_verb.lower() == 'exit':
            result = (round((right_answers/total)*100, 1))
            break
    elif input_verb != res_2form[random_verb]:
        print('You are wrong. Please, try again!', f'{random_verb} --> you wrote - {input_verb}, the correct form - {res_2form[random_verb]}', sep = '\n')
    total += 1   

wrong_answers = total-right_answers

print(f'The result is {result}% of correct answers')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('Statistics of correct and incorrect answers')
ax1.bar(['Right answers', 'Wrong answers'], [right_answers, wrong_answers], color ='darkblue', linewidth = 0)
ax2.pie([right_answers, wrong_answers], labels = ['Right answers', 'Wrong answers'],\
         colors = ['royalblue', 'aqua'], autopct='%1.2f%%', explode=[0.05, 0.05])

plt.show()
```
